# Perceptron algorithm
For instance,
1. Take two inputs x1 and x2.
2. On the way to neuron inputs are multiplied by their own weight.
3. If the sum of these input*weight exceed limit, we pass 1.
4. If not, we pass 0.
5. This limit is called threshold(symbol Θ).

* If weight is larger than other inputs, it means the input is more important.

# AND Gate
* has two inputs and one output.
* outcome is 1 only if two inputs are 1.
* To express this gate by using perceptron, we need to set w1, w2, Θ first.


# NAND Gate & OR Gate
* NAND means NotAND and its output is reverse AND gate.
* SO, outcome is 1 only if two inputs are 0.
* Reverse AND gate's paramters and it will work for NAND gate.

<br>
* Human set perceptron's parameter not Computer.

```python
def AND(x1, x2):
  w1, w2, theta = 0.5, 0.5, 0.7
  # Total of input*weight
  tmp = x1*w1 + x2*w2
  if tmp <= theta:
    return 0
  elif tmp > theta:
    return 1

AND(0,0)
AND(1,0)
AND(0,1)
AND(1,1)
```

# Weight & Bias
* Perceptron sums up input sinal * weight + bias
* If the value is over 0, print 1.
* If not, print 0.

```python
import numpy as np
x = np.array([0,1]) # input
w = np.array([0.5, 0.5]) # weight
b = -0.7 # bias
print(w*x)
print(np.sum(w*x))
print(np.sum(w*x) + b)
```

```python
def AND(x1, x2):
  x = np.array([x1, x2])
  w = np.array([0.5, 0.5])
  b = -0.7
  tmp = np.sum(w*x) + b
  if tmp <= 0:
    return 0
  else: return 1

# bias is different from weight in a sense that weight adjust how input affects result and bias how easily neruon is activated.
# If bias is -0.1, neruon is activated when sum of input*weight exceeds 0.1.
# But if bias is -20.0, neruon is activated when sum of input*weight exceeds 20.0.
```

```python
def NAND(x1, x2):
  x = np.array([x1, x2])
  w = np.array([-0.5, -0.5])
  b = 0.7
  tmp = np.sum(w*x) + b
  if tmp <= 0:
    return 0
  else:
    return 1

def OR(x1, x2):
  x = np.array([x1, x2])
  w = np.array([0.5, 0.5])
  b = -0.2
  tmp = np.sum(w*x) + b
  if tmp <= 0:
    return 0
  else:
    return 1
```

# Limit of Perceptron
* Try XOR gate
* SPOILER ALERT! You can not make XOR by using AND, NAND, OR gates.
* Use non-linear other than linear

# multi-layer perceptron
* With multi-layer perceptron, XOR gate is implementable
* In short, by adding one or more layers, we can overcome limits of single layer perceptron.

```python
def XOR(x1,, x2):
  s1 = NAND(x1, x2)
  s2 = OR(x1, x2)
  y = AND(s1, s2)
  return y
```
