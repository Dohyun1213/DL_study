# Chapter 4 Matplotlib
### In this chapter, we will explore data visualization, a technique that makes data more intuitive and easier to understand.
#### we will use 'matplotlib' especially 'pyplot' which is sub module

code below makes labeled line chart on 2D plane

* mpl is alias of matplotlib
* plt is alias of pyplot
* plot(x-value = [0,1,...,n], y-value) : default n is determined by number of y-values, thus x-value is [0,1,2,3] in this case
* plt.xlabel(name of x-axis), plt.ylabel(name of y-axis)
* plt.show(): draw graph


```python
import matplotlib as mpl
import matplotlib.pyplot as plt

plt.plot([1,2,3,4])
plt.ylabel('y label')
plt.xlabel('x label')
plt.show()
```

# plot()     

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-20,20) # from -20 to 20 create integers by step 1
y1 = 2 * x              # 2 * x   y1 
y2 = (1/3) * x ** 2 + 5 # (1/3) * x ** 2 - 5    y2 
y3 =-x ** 2 - 5         # -x ** 2 - 5    y3 

#  ,   ,     
plt.plot(x,y1,'g--', x,y2,'r^-', x, y3, 'b*:')
plt.axis([-30,30,-30,30]) #    
```

# 4.11         
* normal distribution      0  1      .  random    normal   10,000         .  f1  0  1     , f2  2.0  0.5 . f2  f1     3.0          .
* scipy  stats    norm  1,000  ,      p      .
* hist()  density  True         .

```python
import matplotlib.pyplot as plt
import numpy as np

f1 = np.random.normal(loc=0, scale=1, size=10000)
f2 = np.random.normal(loc=3, scale=5, size=10000)

plt.hist(f1, bins=200, color='red',alpha=0.7, label='loc=0, sclae=1' )
plt.hist(f2, bins=200, color='blue', alpha=0.5, label='loc = 3, sclae= 0.5')
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

#  10.0  3 1,000  
data = norm.rvs(10.0, 3, size=1000)
#  
plt.hist(data, bins=20, density=True, alpha=0.6, color='b')

#    
mu, std = norm.fit(data)
# scipy pdf ,     
xmin, xmax = plt.xlim() # x ,  pyplot 
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)
plt.plot(x, p, 'r', linewidth=2)
```
