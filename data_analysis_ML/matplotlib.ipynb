# plot() 함수의 선 그리기 기능들을 알아보자

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-20,20) # from -20 to 20 create integers by step 1
y1 = 2 * x              # 2 * x를 원소로 가지는 y1 함수
y2 = (1/3) * x ** 2 + 5 # (1/3) * x ** 2 - 5 를 원소로 가지는 y2 함수
y3 =-x ** 2 - 5         # -x ** 2 - 5 를 원소로 가지는 y3 함수

# 빨강색 점선, 녹색 실선과 세모기호, 파랑색 별표와 점선으로 함수를 표현
plt.plot(x,y1,'g--', x,y2,'r^-', x, y3, 'b*:')
plt.axis([-30,30,-30,30]) # 그림을 그릴 영역을 지정함
```

* 첫 두 줄은 그림을 그리기 위한 pyplot 서브 모듈을 가져오는 작업이며, 마찬가지로 넘파이를 np라는 별칭으로 사용하는 작업이다.
* 다음 코드는 넘파이에서 지정된 구간에 대하여 수의 시퀀스를 생성하여 다차원배열 x에 할당하는 코드이다.
* 그 아래의 코드 역시 y1,y2,y3 함수를 정의하는 내용이다.
* 그 다음의 코드는 y1,y2,y3 함수를 각각 다른 스타일로 그리는 코드이다.
  * parameter 'g--' 는 green & 점선(--) 형태로 함수를 그릴 것을 기술하고 있다.
  * parmeter 'r^-' 는 red & 실선(-) & 세모 모양의 표식(^) 형태로 함수를 그릴 것을 기술하고 있다.
  * parameter 'b*:' 는 blue & 점선(:) & 별표 모양의 표식(*) 형태로 함수를 그릴 것을 기술하고 있다.

* matplotlib.org 홈페이지에 다양한 표식에 대한 표가 정리되어 있으므로 참고하자.

# 4.5 제목과 레이블, 스타일에 대해 알아보자
앞 절의 그림에 제목과 범례 등을 넣어서 그림을 좀 더 이해하기 쉽게 만드는 방법을 알아보자. 그림의 제목은 plt.title()함수로 만들 수 있으며, plt.xlable(),  plt.ylable()로 x 축, y축에 레이블링하는 것도 가능하다. 또한, legend() 함수를 통해서 범례를 추가하는 것도 가능하다.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, np.pi * 2, 100)

plt.title('Sin cos curve')
plt.plot(x, np.sin(x), 'r-', label = 'sin curve')
plt.plot(x, np.cos(x), 'b:', label = 'cos curve')
plt.xlabel('x value')
plt.ylabel('y value')
plt.legend() # 범례 추가
```

# 4.6 Figure, axes에 대하여 살펴보자
matplotlib의 pyplot 서브 모듈의 plot() 함수를 사용해서 그림을 그리는 것은 간단한 그림을 빠르게 그리는데 편리하기는 하지만 다양한 종류의 그래프를 한 화면에 여러 개 그리기에는 부족하다. 이 기능을 사용하기 위해서는 pyplot 모듈의 subplots() 함수를 사용해야한다. 이 함수는 다음과 같이 사용하며, 주로 필요한 행의 수와 열의 수를 입력한다.
<br> fig, ax = matplotlib.pyplot.subplots(nrows=2, ncols=2, ....)<br>
subplots() 함수는 Figure 객체와 Axes 객체를 반환하는데 Figure 객체는 축과 그래픽, 텍스트 등의 그리기 객체들을 포함하는 컨테이너 객체이며, Axes 객체는 그림을 그릴 수 있는 테두리 상자 모양의 공간이다.
<br> 예를 들어 행의 수와 열의 수가 모두 2일 경우 2x2 크기의 그림을 그릴 수 있는 공간이 만들어지며 이 공간상에 필요한 그림을 그릴 수 있다. 이때, ax 반환값에 그 공간이 반환되므로 ax는 인덱싱을 통해서 필요한 그림을 그릴 수 있다. 좀 더 구체적으로 설명하자면, ax는 다차원 배열 객체이므로 ax[0,0]과 같은 인덱스를 사용하여 특정한 Axes 객체에 그림을 그릴 수 있다. 반면 Axis는 x, y축 객체이니 표기가 유사ㄴ한 Axes와 Axis를 혼동하지 않도록 주의하도록 하자.

<br><br>
아래의 코드에서 subplots() 함수의 인자는 2,2인데 이 인자의 의미는 2행 2열의 그림을 그리는 공간을 생성한다는 의미이다. 이때 반환값 ax에 [0,0], [0, 1], [1,0], [1,1] 인덱스를 줄 수 있으며, 이 인덱스가 바로 행과 열을 지시한다.

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(2,2) # 2행 2열의 그림을 그리는 공간을 만든다

x = np.random.randn(100)
y = np.random.randn(100)  # 정규 분포를 가지는 데이터
ax[0,0].scatter(x, y)     # 산점도 그림

x = np.arange(10)         # 0에서 9 사이의 연속값
y = np.random.uniform(1, 10, 10) # 균일분포값 생성
ax[0,1].bar(x,y)          # 막대 차트 그림

x = np.linspace(0, 10, 100)
y = np.cos(x)
ax[1,0].plot(x,y)         # 실선으로 함수를 그림

z = np.random.uniform(0,1, (5,5))
ax[1,1].imshow(z)         # 분포를 2D 이미지로 그림
```

이 코드의 결과는 그림과 같이 산점도 그림, 막대 차트 그림, 실선 함수 그림, 이차원 이미지와 같이 각기 다른 스타일의 그림이 나타난다. 그림을 그리는 함수와 그 기능은 아래 자세히 설명되어 있다.
* scatter(): 산점도 그림을 그리는 기능으로 2차원 좌표에 흩뿌려진 점들의 좌표로 자료값을 나타냄
* bar(): 막대 그래프를 그리는 기능, 막대의 모양은 수직 막대.
* barh(): 수평 막대 그래프를 그리는 기능.
* pie(): 파이 형태의 그림을 그리는 기능.
* imshow(): 이미지를 화면에 그리는 기능.
* plot(): 실선이나 점선으로 데이터를 나타냄.

# 4.7 subplot()의 고급 기능
다음은 2행 3열의 격자 공간을 만들고 이 공간에 그리기 객체를 배치하는 방법이다. 우선 subplots() 메소드의 인자를 2,3으로 두고 모두 6개의 그리드를 생성하였다. 다음으로 2층 for 루프를 실행하여 ax[i,j]로 표시된 ax 객체 각각에 대하여 (0,0),....,(1,2)의 레이블로 위치를 표시 하였다. 레이블 출력을 위해서는 text() 함수를 사용하였으며 이 함수에 상대 좌표값, 출력할 문자, 글꼴의 크기를 인자로 넣어주었다.

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(2,3)

for i in range(2):
  for j in range(3):
    ax[i,j].text(0.3, 0.5, str((i,j)),fontsize = 11) # only text
```

격자 모양의 공간은 수직이나 수평으로 합병하여 더 큰 그림을 만드는 것도 가능한데, 이를 위해서는 다음과 같이 GridSpec 클래스를 사용한다. 이 클래스는 생성 시에 행과 열의 개수 너비와 높이를 인자로 받아서 grid라는 인스턴스를 만든다..

# 4.8 자료값의 분포를 나타내는 산점도와 막대 그래프
* 산점도는 두 변수의 상관관계를 직교 좌표계의 평면에 데이터를 점으로 표현하는 그래프로 scatter()함수를 사용하면 쉽게 산점도를 그릴 수 있다.
* scatter() 함수의 인자 중에서 s 키워드 인자는 점의 면적, c는 점의 색상, alpha는 투명도를 나타낸다. 점의 면적 s는 포인트 단위로 출력한다. 표식을 의미하는 marker 키워드 인자를 넣지 않으면 디폴트 표식은 원이 되는데, 이 표식을 그림과 같이 사각형(r:rectangle)과 오각형(p:pentagon) 등의 다양한 모양으로 변경할 수도 있다. alpha 인자는 투명도 값으로 0일 경우 완전 투명, 1일 경우 완전 불투명이며 그 사이의 값으로 자유롭게 설정할 수 있다.

# 4.9 파이 차트와 히트맵 표현
범주형 데이터들이 가지는 전체적인 비율을 한눈에 보고자 할 때는 막대 그래프나 산점도 그래프보다 파이 형태의 그래프가 매우 적절할 것이다.
* pie() 함수를 사용하여 파이 차트를 그릴 수 있으며 함수의 인자인 autopct는 부채꼴 안에 표시될 숫자의 형식을 지정하며 %.2f 값은 소수점 아래 둘째 자리까지를 표시하라는 명령이다. 만일 %까지 표시하고자 할 경우 %.2f%%와 같이 %를 두개 연속해서 입력해야만 한다. colors 키워드 인자는 각 섹션의 색상을 지정하는 역할을 하는데 이를 생략하면 디폴트 색상을 부여한다.
* labels 키워드 인자는 영역별 레이블값을 지정하는 역할을 하며, explode는 중신에서 벗어나는 정도를 백분율로 지정한다.
* heat map이란 열을 의미하는 heat와 지도를 의미하는 map을 결합한 단어로 이미지 위에 열분포 형태로 정보를 표현하는 방법을 의미한다. 열 분포란 실제 지표의 온도를 의미하기도 하지만 데이터의 값이 클 경우와 작은 경우에 대하여 서로 다른 생상을 사용하여 데이터의 분포를 쉽게 이해할 수 있는 유용한 기능이다.
* imshow() 함수 내의 키워드 인자로 color map의 약어인 cmap을 넣으면 'cool', 'Greys'등과 같은 색상 맵 타입을 설정가능

```python
import matplotlib.pyplot as plt
import numpy as np

n = 30                    # 전체 점의 개수
x = np.random.randn(n)    # 점들의 임의의 x 좌표
y = np.random.randn(n)    # 점들의 임의의 y 좌표
colors = np.random.rand(n)# 랜덤한 색상
area = (300 * np.random.rand(n)**2)  # 반지름이 0에서 15 포인트 범위, 커질수록 점 크기가 커짐

plt.scatter(x,y, s = area, c = colors, alpha = 0.5, marker = 'p') # marker parameter를 pentagon으로 설정함
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = [5, 4, 6, 11]
clist = ['cyan', 'gray', 'orange', 'red']
explode = [.06, .07, .08, .09]
plt.pie(data, autopct = '%.2f%%', colors = clist, explode = explode) # 표시될 숫자는 소수점 둘째 자리까지 보여지며 뒤에 %가 붙음, 각 구간의 색은 clist로 정함, explode로 중심에서 벗어나는 정도를 지정함
```

# 4.10 히스토그램
* 히스토그램은 주어진 자료를 몇 개의 구간으로 나누고 각 구간의 도수를 조사하여 나타낸 막대 그래프이다.
* 자료들을 사용하여 히스토그램을 만들어주려면 데이터의 값을 동일한 구간으로 나눠야 하는데 데이터가 들어가는 통을 bin이라고 한다.
* 또 다른 인자인 cumulative는 누적 히스토그램 값을 볼때 매우 유용하다. True 혹은 False를 넣어 자료값들을 누적해서 보여준다.

# 4.11 히스토그램을 이용한 정규 분포 함수와 확률 밀도 함수 그리기
* normal distribution이란 연속 활률 분포의 하나이며 평균이 0이고 분산이 1인 정규 분포를 표준 정규 분포라 한다. 넘파이의 random 서브 모듈에 있는 normal 함수를 이용하여 10,000개의 난수를 생성한 후 이 난수값들을 히스토그램을 사용하여 그려보도록 하자. 이때 f1은 평균이 0이고 분산이 1인 표준 정규 분포 함수라고 하면, f2는 평균이 2.0이고 분산이 0.5가 되도록하자. f2의 분산이 f1보다 작으르모 이 함수는 평균값 3.0 주위로 더 많이 분포하는 성질을 가진다는 것을 알 수 있다.
* scipy 모듈의 stats 서브 모듈에 있는 norm을 사용하여 1,000개의 데이터를 생성하여, 이 데이터들의 확률 밀도 함수 p를 생성하여 그림을 그리는 과정을 보여주고 있다.
* hist() 함수의 density 인자가 True로 설정되어 있으면 히스토그램을 그리면서 동시에 확률 밀도 값을 반환한다.

```python
import matplotlib.pyplot as plt
import numpy as np

f1 = np.random.normal(loc=0, scale=1, size=10000)
f2 = np.random.normal(loc=3, scale=5, size=10000)

plt.hist(f1, bins=200, color='red',alpha=0.7, label='loc=0, sclae=1' )
plt.hist(f2, bins=200, color='blue', alpha=0.5, label='loc = 3, sclae= 0.5')
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

# 평균이 10.0이고 분산이 3인 1,000개의 데이터를 생성
data = norm.rvs(10.0, 3, size=1000)
# 히스토그램을 그림
plt.hist(data, bins=20, density=True, alpha=0.6, color='b')

# 정규 분포를 데이터에 피팅함
mu, std = norm.fit(data)
# scipy의 pdf는 평균, 표준편차로부터 확률 밀도 함수를 생성한다
xmin, xmax = plt.xlim() # x의 최소, 최대값을 pyplot으로부터 구함
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)
plt.plot(x, p, 'r', linewidth=2)
```
